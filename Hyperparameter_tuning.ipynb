```python
import matplotlib.pyplot as plt

# Data
epochs = range(1, 11)
train_loss = [0.2222, 0.0661, 0.0106, 0.0043, 0.0057, 0.0070, 0.0059, 0.0054, 0.0043, 0.0036]
train_accuracy = [0.9068, 0.9780, 0.9979, 0.9991, 0.9986, 0.9979, 0.9983, 0.9984, 0.9987, 0.9991]
val_loss = [0.1481, 0.1605, 0.1847, 0.2350, 0.2875, 0.2868, 0.2952, 0.2815, 0.3546, 0.3397]
val_accuracy = [0.9442, 0.9381, 0.9473, 0.9499, 0.9459, 0.9468, 0.9418, 0.9456, 0.9432, 0.9491]

# Plotting Training and Validation Loss
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, label='Training Loss')
plt.plot(epochs, val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plotting Training and Validation Accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, train_accuracy, label='Training Accuracy')
plt.plot(epochs, val_accuracy, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
# Data
accuracies = [0.9465, 0.9371]
losses = [0.3533, 0.1608]
models = ['First Result', 'Second Result']

# Creating the accuracy plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(models, accuracies, color=['blue', 'green'])
plt.ylim(0.9, 1)
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')

# Creating the loss plot
plt.subplot(1, 2, 2)
plt.bar(models, losses, color=['blue', 'green'])
plt.ylabel('Loss')
plt.title('Model Loss Comparison')

plt.tight_layout()
plt.show()
```

# Comparison:
**Performance:** The first result has a higher accuracy (94.65% vs. 93.71%). However, the second result has a lower loss (0.1608 vs. 0.3533). A lower loss typically indicates the model is better at making precise predictions, but it's not always directly correlated with higher accuracy. It's crucial to consider both metrics when evaluating model performance.

**Efficiency:** The second result took slightly longer per batch (16ms vs. 12ms). This might be due to the different model complexities and batch sizes

```python
import matplotlib.pyplot as plt

# Data for First Model
epochs = range(1, 11)
train_accuracy_first = [0.9068, 0.9780, 0.9979, 0.9991, 0.9986, 0.9979, 0.9983, 0.9984, 0.9987, 0.9991]
val_accuracy_first = [0.9442, 0.9381, 0.9473, 0.9499, 0.9459, 0.9468, 0.9418, 0.9456, 0.9432, 0.9491]

# Data for Second Model
train_accuracy_second = [0.8682, 0.9320, 0.9537, 0.9629]
val_accuracy_second = [0.9249, 0.9371, 0.9392, 0.9418]

# Creating the accuracy plot
plt.figure(figsize=(12, 6))
plt.plot(epochs, train_accuracy_first, label='Training Accuracy (First Model)', color='blue')
plt.plot(epochs, val_accuracy_first, label='Validation Accuracy (First Model)', color='blue', linestyle='dashed')
plt.plot(epochs[:len(train_accuracy_second)], train_accuracy_second, label='Training Accuracy (Second Model)', color='orange')
plt.plot(epochs[:len(val_accuracy_second)], val_accuracy_second, label='Validation Accuracy (Second Model)', color='orange', linestyle='dashed')
plt.title('Accuracy Comparison Between Models')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Data for First Model
train_loss_first = [0.2222, 0.0661, 0.0106, 0.0043, 0.0057, 0.0070, 0.0059, 0.0054, 0.0043, 0.0036]
val_loss_first = [0.1481, 0.1605, 0.1847, 0.2350, 0.2875, 0.2868, 0.2952, 0.2815, 0.3546, 0.3397]

# Data for Second Model
train_loss_second = [0.3278, 0.1999, 0.1430, 0.1140]
val_loss_second = [0.1879, 0.1608, 0.1842, 0.2011]

# Creating the loss plot
plt.figure(figsize=(12, 6))
plt.plot(epochs, train_loss_first, label='Training Loss (First Model)', color='blue')
plt.plot(epochs, val_loss_first, label='Validation Loss (First Model)', color='blue', linestyle='dashed')
plt.plot(epochs[:len(train_loss_second)], train_loss_second, label='Training Loss (Second Model)', color='orange')
plt.plot(epochs[:len(val_loss_second)], val_loss_second, label='Validation Loss (Second Model)', color='orange', linestyle='dashed')
plt.title('Loss Comparison Between Models')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Interpreting the Plots:

##Conclusion on Model Comparison
###Model 1:

###Model 2:

Overall Recommendation:

Model 2 Generalisation Capability: Model 2 appears to generalise better to unseen data. Its performance, characterised by closer training and validation accuracies and a more stable validation loss, suggests it is less prone to overfitting.
Preferred Model for Generalisation: Since the goal is to achieve better generalisation on unseen data, Model 2 seems to be a more suitable choice.
